After successfully mappping the code with name we anonymize the names with some alphanumeric values we create in the next function

```python
# In the below function we are passing "LOVE.ALAYON" and "MIRZA.NAZIR" as list of aliases.
# So, when we apply anonymization function the alphanumeric values for both of them should be equal.
```

```python
list_fields = ['US_CODE']
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 400)

def anonymization(list_fields,df,list_list_aliases=None):
    try:
        if 'xlsx' in df:
            df = pd.read_excel(df)
        elif 'csv' in df:
            df = pd.read_csv(df)
        dict = {}
        # for every column in list_fields
        for fields in list_fields:
            # for every unique row in the field
            for row in df[fields].unique():
                encode = ''.join(random.choices(string.ascii_letters + string.digits, k=8))
                # Adding the key and values to dict for all the columns
                dict[row] = encode

        # Adding key and values for aliases to the same dictionary created and after running all the aliases will have the same encrypted value
        for lst in list_list_aliases:
            tmp1, tmp2 = lst[0], lst[1]
            try:
                dict[tmp2] = dict[tmp1]
            except Exception:
                pass
        print(dict)

        # Columns required for masking
        df3 = df[list_fields]

        for column in df3:
            for key, value in dict.items():
                # for every key and value in the created dictionary, we update the value(encoded) when key matches in the column in the dataframe
                df.loc[df[column] == key, column] = value
        # create a csv output file with encoding
        df.to_csv('out1.csv', sep=',', encoding='utf-8')

    except Exception as error:
        print(error)

anonymization(list_fields,df='df1.csv',list_list_aliases = [['LOVE.ALAYON','MIRZA.NAZIR']])
```

We can see from the above dictionary that both "LOVE.ALAYON" and "MIRZA.NAZIR" share same alphanumeric values

We will also verify the same in the resulting dataframe.
